```python
data=list(df["products"].apply(lambda x:x.split(",")))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
df1 = apriori(df,min_support=0.1,use_colnames=True)
df1.head()
```

```python
df1.shape
```

```python
df1.sort_values(by = 'support', ascending = False)
```

```python
df1['length'] = df1['itemsets'].apply(lambda x:len(x))
df1
```

```python
df1[(df1['length']==2) & (df1['support']>=0.1)]
```

```python
report=association_rules(df1, metric = "confidence", min_threshold = 0.2).sort_values(by='confidence', ascending=False)
report
```
